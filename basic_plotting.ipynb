```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
plt.style.use('seaborn-whitegrid')
```

```python
x = np.linspace(0, 10, 100)
fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
```

```python
fig, ax = plt.subplots(2)
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

```python
# start by creating a figure and an axes

fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x))
```

```python
plt.plot(x, np.sin(x - 0), color='blue')
plt.plot(x, np.sin(x - 1), color='g')
plt.plot(x, np.sin(x-2), color='0.75')
```

```python
plt.plot(x, x+1, linestyle='solid')
plt.plot(x, x+2, linestyle='dashed')
plt.plot(x, x+5, linestyle=':')
```

```python
# setting the axis limits

plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

```python
# labeling your figure

plt.plot(x, np.sin(x))
plt.title('A sine wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
# adding a legend

plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.legend()
```

```python
# scatter plot

x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black')
```

```python
plt.scatter(x, y, marker='o')
```

```python
# creating a histogram

data = np.random.randn(1000)
plt.hist(data)
```

```python
plt.hist(data, bins=30, density=True, alpha=0.3,
         histtype='stepfilled', color='steelblue')
```

```python
# learning to use seaborn

import seaborn as sns
sns.set()
```

```python
rng = np.random.RandomState(0)
x = np.linspace(0, 10, 500)
y = np.cumsum(rng.randn(500, 6), 0)
plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper left')
```

```python
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data1 = pd.DataFrame(data, columns=['x', 'y'])

for col in 'xy':
    plt.hist(data1[col], density=True, alpha=0.5)
```

```python
sns.distplot(data1['x'])
sns.distplot(data1['y'])
```

```python
# exploring correlations with pair plots

iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris, hue='species', height=2.5)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# computing factor plots - multivariate plotting

with sns.axes_style(style='ticks'):
    g = sns.factorplot('day', 'total_bill', 'sex', data=tips, kind='box')
    g.set_axis_labels('Day', 'Total_bills')
```

```python
sns.jointplot('total_bill', 'tip', data=tips, kind='reg')
```

```python
planets = sns.load_dataset('planets')
planets.head()
```

```python
with sns.axes_style('white'):
    g = sns.factorplot('year', data=planets, aspect=2,
                       kind='count', color='steelblue')
    g.set_xticklabels(step=5)
```
